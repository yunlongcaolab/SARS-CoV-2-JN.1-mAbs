```python
import pandas as pd
sources =  [
    'XBB infection',
    'XBB BTI',
    'BA.5 + XBB infection',
    'BA.5 + JN.1 infection',
    'BA.5 BTI + HK.3 infection',
    'BA.5 BTI + XBB infection',
    'BA.5 BTI + JN.1 infection'
]

dms = pd.read_csv("../data/DMS/antibody/_clustering.csv")
abinfo = pd.read_csv("../data/_mAb_info_clean.csv").merge(
    dms[['new_group','id']], on='id', how='outer'
).fillna('N.D.').query(
    'new_group != "N.D." or source in @sources'
)

BD45_map = {
'BD45-1': 'S309',
'BD45-105': 'S2H97',
'BD45-117': 'FC08',
'BD45-14': 'S304'
}

def name_mapping(x: str):
    if x.startswith('BD30-'):
        return 'BD-'+x[5:]
    if x.startswith('BD55-'):
        number = int(x[5:])
        if number < 300:
            return 'BD-'+str(number + 630)
    if x.startswith('GC'):
        return 'BD57-'+x[2:]
    if x.startswith('SN'):
        return 'BD56-'+x[2:]
    if x.startswith('CP'):
        return 'BD58-'+x[2:]
    if x in BD45_map:
        return BD45_map[x]
    return x

show_cols = {
    'id': 'antibody',
    'source': 'source',
    'paper_reactivity': 'reactivity',
    'WT_ELISA_1ug': 'WT RBD ELISA',
    'BA5_ELISA_1ug': 'BA.5 RBD ELISA',
    'XBB1_5_ELISA_1ug': 'XBB.1.5 RBD ELISA',
    'JN1_ELISA_1ug':'JN.1 RBD ELISA',
    'v_gene_H':'VH gene',
    'd_gene_H':'DH gene',
    'j_gene_H':'JH gene',
    'c_gene_H':'H isotype',
    'v_gene_L':'VL gene',
    'j_gene_L':'JL gene',
    'v_domain_shm_nt_H':'VH SHM (nt)',
    'v_domain_shm_nt_L':'VL SHM (nt)',
    'v_domain_shm_ratio_H':'VH SHM ratio',
    'v_domain_shm_ratio_L':'VL SHM ratio',
    'D614G_IC50':'D614G',
    'BA1_IC50':'BA.1',
    'BA2_IC50':'BA.2',
    'BA5_IC50':'BA.5',
    'XBB1_5_IC50':'XBB.1.5',
    'HK3_1_IC50':'HK.3.1',
    'BA2_86_IC50':'BA.2.86',
    'JN1_IC50':'JN.1',
    'JN1_F456L_IC50':'JN.1+F456L',
    'JN1_R346T_F456L_IC50':'KP.2',
    'JN1_F456L_A475V_IC50':'JN.1+F456L+A475V',
    'KP3_IC50':'KP.3',
    'KP3_A475V_IC50':'KP.3+A475V',

    'new_group':'epitope group',
}


abinfo.assign(
    id = lambda x: [name_mapping(y) for y in x['id']])[show_cols.keys()].rename(columns=show_cols).to_csv("../data/_Table_S2.csv", index=None)
```
